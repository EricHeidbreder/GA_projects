```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(houses_train_cleaned[['Overall Qual', 'Overall Qual x Gr Liv Area', 'Gr Liv Area', 'log_gr_liv_area', 'age_of_garage', 'SalePrice']]);
```
