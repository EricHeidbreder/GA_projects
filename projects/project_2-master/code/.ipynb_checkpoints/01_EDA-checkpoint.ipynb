```python
sns.set_style('darkgrid')
plt.scatter(houses_train['1st Flr SF'], houses_train['SalePrice'], alpha=0.5)
plt.scatter(houses_train['Gr Liv Area'], houses_train['SalePrice'], alpha=0.5)
plt.legend(labels=['1st Flr Sf', 'Gr Liv Area'])
plt.title('1st Floor Square Footage and Above Ground Living Area\nSq. Ft vs. Price', fontsize=16)
plt.xlabel('Square Footage')
plt.ylabel('Price');
```

Looks like we have some outliers in a few of the Square Footage categories

```python
sns.set_style('darkgrid')
sns.scatterplot(x=houses_train['Garage Yr Blt'],
                  y=houses_train['SalePrice'],
                  alpha=0.5)
plt.title('Garage Year Built vs. Sale Price', fontsize=16);
```

Garages can't be built in the future, so we'll remove that row. Also, it looks like garages built recently have a drastically higher price than those built before the year 2000.

```python
houses_train = houses_train[houses_train['Garage Yr Blt'] != 2207]
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(9, 2))
sns.boxplot(houses_train['SalePrice'])
plt.title('Distribution of Sales Prices', fontsize=20)
plt.tight_layout()
plt.savefig('../images/sales_price_boxplot', pad_inches=0);
```

The interquartile range of Sales Prices is between 129825 - 214000, so that's a very small range with a lot of data

```python
houses_train.groupby('Year Built').mean()['SalePrice'].plot(kind='bar',
                                                                    figsize=(35,5))
plt.title('Sale Price by Year Built', fontsize=16);
```

I don't want to use the year as it is, I'm going to convert it to age later in project (*See 02_Cleaning, Engineering*)

```python
greater_than_1982 = houses_train['Year Built'] > 1982
x = ['Pre-1983', '1983 to 2010']
y = [houses_train[~greater_than_1982].mean()['SalePrice'], houses_train[greater_than_1982].mean()['SalePrice']]
plt.bar(data=houses_train, x=x, height=y, color='darkgreen')
plt.title('Mean sale prices for homes built \nbefore 1983 and post 1983', fontsize=16)
plt.savefig('../images/home_price_pre_post_1983');
```

After 1984, the mean house price seems to jump up significantly from ~140000 to ~238000

### Features that could benefit from log transformation

```python
columns_to_log = ['BsmtFin SF 1', 'Lot Area']
houses_train[columns_to_log].hist(figsize=(12,4))
plt.tight_layout();
```

These columns are all right-skewed, so logarithmic transformation should balance these out a bit to make them more useful features

```python
houses_train.groupby('Exter Qual').mean()['SalePrice'] # External Quality seems to have a significant impact on SalePrice
```

```python
ovr_qual_saleprice = houses_train.groupby('Overall Qual').mean()['SalePrice']

sns.barplot(ovr_qual_saleprice.index, ovr_qual_saleprice.values)
plt.title('Mean Price for Overall Quality Rating', fontsize=16)
plt.xlabel('Overall Quality')
plt.ylabel('Price')
plt.savefig('../images/mean_price_overall_quality.png');
```

There seems to be a logarithmic shape to the mean price for quality score. I wonder if this is reflected in more of the data?
